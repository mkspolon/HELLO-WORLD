```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.linspace(-1,1,5)
y2 = np.linspace(-3,3,5)
x = np.arange(5)
```

```python
fig = plt.figure()
ax = fig.subplots()
line1 = ax.plot(x,y, label='a')
line2 = ax.plot(x,y2, label='b')
ax.legend() #adds a legend
ax.set_title('Random data')
ax.set_ylabel('date')
ax.set_xlabel('height')

ax.axvline(2,ymin=0,ymax=1,color='red',linestyle='--',zorder=1)
ax.axhline(0,color='#FF0099',linestyle=':',xmin=0,xmax=1)
```

```python
fig2 = plt.figure()
ax2 = fig2.subplots(2,2)
plt.tight_layout()
ax2[0,0].plot(x,y2,color='red',label='2010')
ax2[0,0].plot(x,y,color='green',label='2011')
ax2[0,1].scatter(x,y2)
ax2[1,0].bar(x,y2)
ax2[1,1].barh(x,y2)
ax2[0,0].legend()
```

```python
people = ['student_a','student_b']
studentA = np.array([90,23,75,45])
studentB = np.array([80,63,71,42])

x = np.arange(len(studentA))

fig,ax = plt.subplots()
width=0.3
bar1 = ax.bar(x-width/2, studentA, width)
bar2 = ax.bar(x+width/2, studentB, width, color='green')
ax.set_ylim([0,100])
ax.set_title('student grades')
ax.set_xlabel('test num')
ax.set_ylabel('grade')
ax.set_xticks([0,1,2,3])
ax.set_xticklabels(["ex1","ex2","ex3","ex4"])
ax.legend(people)

for i in range(len(studentA)):
    bar1[i].set_linewidth(3.5)
    bar2[i].set_linewidth(3.5)
    if studentA[i]<50:
        bar1[i].set_edgecolor('red')
    elif studentB[i]<50:
        bar2[i].set_edgecolor('red')
```
