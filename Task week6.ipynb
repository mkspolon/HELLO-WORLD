```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

your_dataset=pd.read_csv('Churn_Modelling.csv')

# just add to this list all the columns you want to scale
VariableList = [ 'Age',"CreditScore"]

scaler = MinMaxScaler()

# removed second loop for optimization
# iterate over variables in VariableList
for j in range(0,len(VariableList)):
    # extract target column
    temp = your_dataset[VariableList[j]]
    # reshape to a df as long as dataset with only 1 column
    temp1 = temp.values.reshape(-1,1)
    # fit scaler
    scaler.fit(temp1)
    # scale to a range
    norm = scaler.transform(temp1)
# ----------------------IMPORTANT----------------------
    # convert your_dataset column into float*********
    # This is where our problem was coming from
    # It was converting our range eg >> 0.37 into an int: 0
    your_dataset[VariableList[j]] = norm.astype("float")
# ----------------------IMPORTANT----------------------
your_dataset
```
